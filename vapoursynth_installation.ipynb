##### * Add shared libraries(.so files) to path. This is for avoiding libvapoursynth.so import error.

```python
%cd /content/
```

```python
import os
default_ld_path = "/usr/local/lib"
ld_env = os.environ.get('LD_LIBRARY_PATH', '')
if default_ld_path not in ld_env:
    new_ld_env = f"/usr/local/lib:{os.environ.get('LD_LIBRARY_PATH', '')}"
else:
    new_ld_env = ld_env
```

```python
%env LD_LIBRARY_PATH=$new_ld_env
```
